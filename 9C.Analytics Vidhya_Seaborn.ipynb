```python
#Importing libraries
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
#Seaborn Themes Setting
sns.set()
sns.set(style = 'ticks')
```

```python
#Reading excel file
data = pd.read_csv("https://raw.githubusercontent.com/akki8087/Big-Mart-Sales/master/Train.csv")
data.head()
```

```python
#Droping Null Value
data = data.dropna(how = 'any')
```

```python
#Creating new variable Visibility
data['Visibility_scaled'] = data['Item_Visibility']*100
```

```python
#View the column names
data.columns
```

```python
#View the first five rows of the data
data.head()
```

```python
#Line Plot
sns.lineplot(x = 'Item_Weight', 
             y = 'Item_MRP', 
             data = data[:50]);
```

```python
#Bar Plot
sns.barplot(x = 'Item_Weight', 
            y = 'Item_MRP', 
            data = data[:5]);
```

```python
#Distribution Plot
sns.distplot(data['Item_MRP']);
```

```python
#Box Plot -Vertical
sns.boxplot(data['Item_Outlet_Sales'], 
            orient = 'vertical');
```

```python
#Box Plot - Horizontal
sns.boxplot(data['Item_Outlet_Sales'], 
            orient = 'horizontal');
```

```python
#Violin Plot
sns.violinplot(data['Item_Outlet_Sales'], 
               orient = 'vertical', 
               color = 'magenta');
```

```python
#Scatter Plot
sns.relplot(x = 'Item_MRP', 
            y = 'Item_Outlet_Sales', 
            data = data[:200], 
            kind = 'scatter');
```

```python
#Scatter Plot with Category wise color
sns.relplot(x = 'Item_MRP', 
            y = 'Item_Outlet_Sales', 
            hue = 'Item_Type', 
            data = data[:200]);
```

```python
#Line Plot with Category wise color
sns.lineplot(x = 'Item_Weight', 
             y = 'Item_MRP', 
             hue = 'Outlet_Size', 
             data = data[:150]);
```

```python
#Bubble Plot
sns.relplot(x = 'Item_MRP', 
            y = 'Item_Outlet_Sales', 
            kind = 'scatter', 
            size = 'Visibility_scaled', 
            hue = 'Visibility_scaled', 
            data = data[:200]);
```

```python
#Category Wise Sub Plot
sns.relplot(x = 'Item_Weight', 
            y = 'Item_Visibility',
            hue = 'Outlet_Size',
            col = 'Outlet_Size', 
            data = data[:100]);
```

```python
#Strip Category Wise Plot
sns.catplot(x = 'Outlet_Size', 
            y = 'Item_Outlet_Sales', 
            kind = 'strip', 
            data = data[:250]);
```

```python
#Swarm Category Wise Plot
sns.catplot(x = 'Outlet_Size', 
            y = 'Item_Outlet_Sales', 
            kind = 'swarm', 
            data = data[:250]);
```

```python
#Box Category Wise Plot
sns.catplot(x = 'Outlet_Size', 
            y = 'Item_Outlet_Sales', 
            kind = 'box', 
            data = data[:250]);
```

```python
#Violin Category Wise Plot
sns.catplot(x = 'Outlet_Size', 
            y = 'Item_Outlet_Sales', 
            kind = 'violin', 
            data = data[:250]);
```

```python
#Bar Category Wise Plot
sns.catplot(x = 'Outlet_Size', 
            y = 'Item_Outlet_Sales', 
            kind = 'bar', 
            data = data[:250]);
```

```python
#Setting the size of the plot
plt.figure(figsize = (7,7))

#Density Plot
sns.kdeplot(data['Item_Visibility'], 
            shade = True);
```

```python
#Setting the size of the plot
plt.figure(figsize = (7,7))

#Density Plot
sns.kdeplot(data['Item_MRP'], 
            shade = True);
```

```python
#Setting the size of the plot
plt.figure(figsize = (7,7))

#DIstribution Plot with Density
sns.distplot(data['Item_Outlet_Sales']);
```
