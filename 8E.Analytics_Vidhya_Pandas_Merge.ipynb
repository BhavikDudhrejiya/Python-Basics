# Concate the dataframe

```python
#Importing library
import pandas as pd
```

```python
#Creating 3 data frame
data_Frame1 = pd.DataFrame({'A':[10,20,30,40,50,60], 
                            'B':[70,80,90,100,110,120]},
                           index = [0,1,2,3,4,5])

data_Frame2 = pd.DataFrame({'A':[60,50,40,30,20,10], 
                            'B':[120,110,100,90,80,70]},
                           index = [6,7,8,9,10,11])

data_Frame3 = pd.DataFrame({'A':[130,140,150,160,170,180], 
                            'B':[190,200,210,220,230,240]},
                           index = [12,13,14,15,16,17])
```

```python
#Printing the dataframe
print(data_Frame1)
print(data_Frame2)
print(data_Frame3)
```

```python
#Concate all the data frame (by default it is a column wise concate)
result = pd.concat([data_Frame1,data_Frame2,data_Frame3])
result
```

```python
#Concate all the data frame with keys
result = pd.concat([data_Frame1,data_Frame2,data_Frame3], keys =['x','y','z'])
result
```

```python
#Accessing the data frames with keys
print(result.loc['y'])
print(result.loc['x'])
print(result.loc['z'])
```

```python
#Creating anoher data frame
data_Frame4 = pd.DataFrame({'A':[210,240,250,260,270,280], 
                            'B':[390,300,310,320,330,340]},
                           index = [1,3,6,8,12,17])

data_Frame4
```

```python
#Concatenating row wise dataframes
result1 = pd.concat([data_Frame1,data_Frame4], axis = 1, sort = False)
print(result1)
result2 = pd.concat([data_Frame2,data_Frame4], axis = 1, sort = False)
print(result2)
result3 = pd.concat([data_Frame3,data_Frame4], axis = 1, sort = False)
print(result3)
```

```python
#Concate all the data frame with keys
result = pd.concat([data_Frame1,data_Frame2,data_Frame3, data_Frame4], 
                   keys =['x','y','z',"p"])
result
```

```python
#Using inner join while concatenate
result1 = pd.concat([data_Frame1,data_Frame4], axis = 1, join = 'inner')
print(result1)
```

```python
#Using inner outer while concatenate
result4 = pd.concat([data_Frame1,data_Frame4], axis = 1, join = 'outer')
print(result4)
```

# Merge dataframe

```python
#Creating 3 data frame
df_a = pd.DataFrame({
    'Subject id':['1','2','3'],
    'first_name':['Alex','Amy','Allen'],
    'last_name':['Anderson','Ackrman','Ali']})

df_b = pd.DataFrame({
    'Subject id':['2','3','6'],
    'first_name':['Billy','Brian','Bryce'],
    'last_name':['Bonder','Black','Bruce']})

df_c = pd.DataFrame({
    'Subject id':['1','2','3','4','5','6'],
    'test id':[51,15,15,61,16,14]})
```

```python
print(df_a)
print('-----------------------------------')
print(df_b)
print('-----------------------------------')
print(df_c)
```

```python
#Merging df_a with df_c
pd.merge(df_a,df_c,on = 'Subject id')
```

```python
#Merging df_b with df_c
pd.merge(df_b,df_c,
         on = 'Subject id')
```

```python
#Merging df_a with df_b with outer (By default it is inner)
pd.merge(df_a,df_b,
         on = 'Subject id')
```

```python
#Merging df_a with df_b with outer
pd.merge(df_a,df_b,
         on = 'Subject id',
         how = 'outer')
```

```python
#Merging df_a with df_b with inner
pd.merge(df_a,df_b,
         on = 'Subject id',
         how = 'inner')
```

```python
#Merging df_a with df_b with right
pd.merge(df_a,df_b,
         on = 'Subject id',
         how = 'right')
```

```python
#Merging df_a with df_b with left
pd.merge(df_a,df_b,
         on = 'Subject id',
         how = 'left')
```
